**Questão 1:**

Crie uma função genérica para retornar os números pares de um array.

```python
def pares(lista):
  pares = [] # vazio para interar aqui

  for n in lista:
    if n % 2 == 0:
      pares.append(n)

  return f'Numeros pares retornados: {pares}'
```

```python
lista_teste = np.arange(10)
pares(lista_teste)
```

**Questão 3:**

Escreva um programa em Python que, dado um número n, imprima uma pirâmide numérica com n linhas.

```python
def piramide():
  ''' faz uma piramide com o numero de n linhas'''
  try:
    n = int(input('Digite um numero inteiro:\t'))
  except ValueError:
    print('Digite apenas numeros pares.')

  for i in range(1, n+1):
    for linha in range(1, i+1):
      print(linha, end=' ')
    print()
```

```python
piramide()
```

**questão 4:**

Crie uma função que eu passe um CEP e me retorne a rua, bairro e cidade.

Dica - utilize: viacep.com.br/ws/01001000/json/

```python
def busca_cep(cep):
  ''' utiliza a api do viacep para retornar: RUA BAIRRO CIDADE '''
  url = f"https://viacep.com.br/ws/{cep}/json/"

  try:
        response = requests.get(url)

        dados = response.json()

        return {
            "rua": dados.get("logradouro"),
            "bairro": dados.get("bairro"),
            "cidade": dados.get("localidade")
        }

  except requests.exceptions.HTTPError as erro_http:
        print("Erro HTTP:", erro_http.response.status_code)
```

```python
busca_cep("01001000")
```
